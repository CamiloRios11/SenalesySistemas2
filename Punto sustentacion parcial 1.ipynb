```python
from numpy import exp, pi, arange, real, imag, abs, angle, sqrt, cos, sin
import matplotlib.pyplot as plt
T = 7*pi/12
t = arange(0, 3*T,T )
plt.figure(1, (25, 10))
x = 10*cos(6*t) - 5*sin(8*t)
plt.plot(t, x)
plt.grid()
plt.margins(x=0)
plt.xticks([0, 12*pi, 24*pi, 36*pi, 48*pi, 60*pi, 72*pi]
           , [0, '$12\pi$', '$24\pi$', '$36\pi$', '$48\pi$', '$60\pi$', '$72\pi$'])
plt.title('Gráfica señal $x(t)=cos(t/3)+cos(t/4)$', fontsize=15)
plt.show()

```
